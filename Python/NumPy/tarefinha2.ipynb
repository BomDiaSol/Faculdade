```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matriz = np.random.randint(10, 99, size =(6,6))
print(matriz)
```

```python
matriz2 = matriz.copy()
matriz2[matriz2 % 2 != 0] = -1
print(matriz2)

# OUTRO MÉTODO

# matriz_modificada = np.where(matriz % 2 != 0, -1, matriz)
# print("\nMatriz Modificada:\n", matriz_modificada)
```

```python
min = np.min(matriz)
max = np.max(matriz)

print(min)
print(max)
```

```python
soma_coluna = np.sum(matriz, axis=0) # Axis 0 pra colunas
soma_linhas = np.sum(matriz, axis=1) # Axis 1 pra linhas

print(soma_coluna)
print(soma_linhas)
```

```python
plt.figure(figsize=(12, 5))

# MATRIZ ORIGINAL

plt.subplot(1, 2, 1)
plt.hist(matriz.flatten(), bins=10, color='blue', edgecolor='black')
plt.title("Histograma - Matriz Original")
plt.xlabel("Valor")
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()
```
